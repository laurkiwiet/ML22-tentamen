```python
import tensorboard as tb

experiment_id = "q794S5URQFWPeZbeMK8hBg"
experiment = tb.data.experimental.ExperimentFromDev(experiment_id)
df_exp = experiment.get_scalars()

```

```python
import pandas as pd
pd.set_option('display.max_rows', None)
# Filter runs ik heb gebruikt voor de opdracht
rows_to_include = "20230125"
df_runs = df_exp[df_exp['run'].str.contains(rows_to_include)]

# De waarden van de values onder de juiste metric plaatsen.
df_runs_clean = pd.DataFrame(df_runs.pivot(index=['run', 'step'], columns=['tag'], values=['value']))

df_runs_clean.reset_index(inplace=True)

df_runs_clean
```
