```python
experiment_id_hypertuning = "dpXbv4Z9QEydxUnuRSaJJA"
experiment_hypertuning = tb.data.experimental.ExperimentFromDev(experiment_id_hypertuning)
df_exp_hypertuning= experiment_hypertuning.get_scalars()
```

```python
import pandas as pd

pd.set_option("display.max_rows", None)
# Filter runs ik heb gebruikt voor de opdracht
best_run = "train_e57bcea9"
df_runs_hypertuning = df_exp_hypertuning[df_exp_hypertuning["run"].str.contains(best_run)]

# De waarden van de values onder de juiste metric plaatsen.
df_runs_clean_hypertuning = pd.DataFrame(
    df_runs_hypertuning.pivot(index=["run", "step"], columns=["tag"], values=["value"])
)

df_runs_clean_hypertuning.reset_index(inplace=True)

df_runs_clean_hypertuning
```

```python
import plotly.express as px
import plotly.figure_factory as ff
```

```python
import pandas as pd
plt_data = pd.DataFrame.copy(df_runs_clean_hypertuning)
plt_data['Epoch'] = plt_data['step']
plt_data['Accuracy'] = plt_data.value['ray/tune/Accuracy']
plt_data['Test loss'] = plt_data.value['ray/tune/test_loss']
#plt_data_totaal = pd.DataFrame(plt_data['Epoch','Accuracy','Test loss'])

#['Epoch', 'Accuracy', 'Test loss']
```

```python
plt_data
```

```python
# Create Line plot
fig = px.line(plt_data, x=plt_data['Epoch'], y=plt_data['Accuracy'])
#fig.add_scatter(plt_data, x=plt_data['Epoch'], y=plt_data['Test loss'])
# Add Scatter plot

fig

from plotly.subplots import make_subplots
import plotly.graph_objects as go
import numpy as np


tickvals = np.arange(plt_data['Epoch'].min(), plt_data['Epoch'].max() + 1)
fig = make_subplots(rows=2, cols=1)

fig.append_trace(go.Scatter(
  x=plt_data['Epoch'], y=plt_data['Accuracy']), row=1, col=1)

max_accuracy = plt_data['Accuracy'].max()
max_accuracy_epoch = plt_data[plt_data['Accuracy'] == max_accuracy]['Epoch'].values


for epoch in max_accuracy_epoch:
  fig.add_annotation(x=epoch,
                    y=max_accuracy,
                    text=round(max_accuracy, 2))

fig.add_trace(go.Scatter(
    x=plt_data['Epoch'], y=plt_data['Test loss']
), row=2, col=1)

min_loss = plt_data['Test loss'].min()
min_loss_epoch = plt_data[plt_data['Test loss'] == min_loss]['Epoch'].values

for epoch in min_loss_epoch:
  fig.add_annotation(x=epoch,
                    y=min_loss,
                    xref='x2',
                    yref='y2',
                    text=round(min_loss, 2))

fig.update_layout(height=600, width=1000, title_text="Accuracy & min loss per epoch", xaxis=dict(tickmode="array", tickvals=tickvals,
        range=[plt_data['Epoch'].min(), plt_data['Epoch'].max()]), xaxis2=dict(tickmode="array", tickvals=tickvals,
        range=[plt_data['Epoch'].min(), plt_data['Epoch'].max()]), xaxis_title="Epoch", xaxis2_title = "Epoch", yaxis_title ="Accuracy", yaxis2_title="Test loss")


fig.show()
```
